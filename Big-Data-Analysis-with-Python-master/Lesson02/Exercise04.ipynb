## Exercise 4: Creating Line Graphs Using Different Libraries

```python
%matplotlib inline
import numpy as np
X = np.arange(0,100)
Y = np.random.randint(0,200, size=X.shape[0])
```

```python
import matplotlib.pyplot as plt
plt.plot(X, Y)
```

```python
import pandas as pd
df = pd.DataFrame({'x':X, 'y_col':Y})
```

```python
plt.plot('x', 'y_col', data=df)
```

```python
df.plot('x', 'y_col')
```

```python
import seaborn as sns
sns.lineplot(X, Y)
sns.lineplot('x', 'y_col', data=df)
```
